## Procedure to create environment

1. To create a conda environment, in the terminal, run `conda create -n myenv python=3.9 scipy numpy` where you can specify the python version and the relevant  packages (with their versions)
1. To see the list of environments you can type `conda env list`
1. To activate an environment, type `conda activate myenv`
1. To get a list of packages in an activated environment, type `conda list`
1. To install more packages in an activated environment, type `pip install networkx`
1. Latest versions of matplotlib require extra dependencies which can be installed by `sudo apt-get install dvipng texlive-latex-extra texlive-fonts-recommended cm-super`
1. To remove an environment, `conda remove --name myenv --all` or `conda remove -p path/to/myenv --all`
