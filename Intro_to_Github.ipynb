# List of commands

Assuming you have already signed in on github, you won't be asked to authenticate. You may need to create personal access tokens for each repository.

Cloning and working on a cloned repository:

1. To clone a repository `git clone https://github.com/Pickett05/Simplicial_Complexes.git`
1. Switch to the directory on your local system `cd Simplicial_Complexes`
1. Initialize git `git init`
1. Tell git who you are using `git config --global user.name "Pickett05"` and `git config --global user.email "nidhisonwane1115@gmail.com"`
1. Connect the remote system to the local branch `git remote add master https://github.com/Pickett05/Simplicial_Complexes.git` where "master" is the name of the local branch (I think)
1. Check status of file changes `git status`
1. Add files to be committed to the current branch `git add README.md`, for all the files run `git add .` and for including folders run `git add *`
1. Commit the changes to the current branch `git commit -m "minor edit"`
1. Push the changes to the remote server using `git push`

Some terms:
1. `git pull` helps the local system get up to date with remote (repo). If your branch is behind, then it will fast forward to the remote server else stuff happens you need to check
1. local branch - your computer
1. remote - server on internet, like GitHub

Working with branches:

1. Check which branch you are currently on `git branch`
1. Check whether the current branch is up to date with repo branch `git checkout`
1. Create a new branch `git branch "sidekick"`
1. Switch to another branch `git checkout sidekick`
1. To push changes to a different branch `git push --set-upstream origin sidekick`

Merging branches:

1. To merge, run `git merge sidekick` in the branch where you want to merge "sidekick"
1. Pull the request from repo and check for conflicts and merge
1. Delete the sidekick branch if required

Modifying files:

1. Whenever we add some files and folders using git, only addition takes place using the command above. Hence, if we have deleted something, the deletion won't reflect in the remote server.
1. We have different flag with `git add` that adds, modifies, and removes index entries to match the working tree: `git add . -A`. This works for all files and folders
1. Then commit: `git commit -m "Deleted some files..."`
1. And push: `git push`

Linking files and folders present in another directory to be pushed in github:
1. Create a hard link to the file that is placed in another directory
1. You can use `ln` command (`mklink` on Windows) to achieve this (assuming that you are currently in the repository directory): `
ln ../../filename` (this will add a file with the same name to the current directory and the same inode)
1. Then add `git add filename` and commit `git commit -m "add file from another directory"`
